Решение первой задачи:
Результат: [10]

```python
A = [6, -8, 3]
B = [1, 1, 4]
alen = len(A)
blen = len(B)
rezult = 0

for i in range(0, alen):
    if i != 0:
        print(' + ', end='')

    print(A[i], end='')
    print('*', end='')
    print(B[i], end='')
    row = A[i] * B[i]
    rezult = rezult + row

print(' = ', end='')
print(rezult, end='')
```

Решение второй задачи:
Результат:  [[  -4,  -13],
            [ 104, -102],
            [ -24,   12],
            [  -4,   -3]]

```python
arr1 = [-3, -7]
arr2 = [-10, 6]
arr3 = [0, -6]
arr4 = [-1, -3]

C = [arr1, arr2, arr3, arr4]

arr5 = [-8, 9]
arr6 = [4, -2]

D = [arr5, arr6]

clen = len(C)
rezult = 0
rezultList = []

for i in range(0, clen):
    CI= C[i]
    cleni = len(C[i])
    innerList = []

    for k in range(0, cleni):
        dlen = len(D)

        for g in range(0, dlen):

            if g != 0:
                print(' + ', end='')

            print(CI[g], end='')
            print('*', end='')
            print(D[g][k], end='')
            row = CI[g] * D[g][k]
            rezult = rezult + row

        innerList.append(rezult)
        print(' = ', end='')
        print(rezult)
        rezult = 0

    rezultList.append(innerList)
    rezult = 0

print()
print('Rezult:')
print(rezultList)
```
