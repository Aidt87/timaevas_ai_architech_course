```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv")

##Количество пустых ячеек
print('Количество пустых ячеек: ')
print(df.isna().sum())

##Заполняем пустые ячейки
print()
print('Заполняем пустые ячейки..... ')
df.loc[df['Cabin'].isna(), 'Cabin'] = 'NotFound'
df.loc[df['Age'].isna(), 'Age'] = df['Age'].max()/2
df.loc[df['Embarked'].isna(), 'Embarked'] = 'N'
print()

##Количество пустых ячеек после заполнения
print('Количество пустых ячеек после заполнения: ')
print(df.isna().sum())

##Аналитика
print()
print('Аналитика: ')
print(' 1. Количество уникальных значений')
print(df.nunique())

print()
print(' 2.1 Минимальное значение')
print(df[["PassengerId", "Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]].min())

print()
print(' 2.2 Максимальное значение')
print(df[["PassengerId", "Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]].max())
```
