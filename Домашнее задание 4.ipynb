Задача №1

```python
import numpy as np
A = np.array([[1, 2, 1],
             [-2, 1, 1]])

B = np.array([[3, 1],
             [-1, 0],
             [2, 4]])

C = np.array([[-2, 2],
             [1, -1],
             [1, 2]])

print(A.T + 2*B)
print(B + 2*C)
```
