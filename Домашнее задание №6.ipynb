```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv")

#Количество пустых ячеек
#print('Количество пустых ячеек: ')
#print(df.isna().sum())

##Заполняем пустые ячейки
print()
print('Заполняем пустые ячейки..... ')
#df.loc[df['Cabin'].isna(), 'Cabin'] = 'NotFound'
#df.loc[df['Age'].isna(), 'Age'] = df['Age'].max()/2
#df.loc[df['Embarked'].isna(), 'Embarked'] = 'N'
#print()

##Количество пустых ячеек после заполнения
#print('Количество пустых ячеек после заполнения: ')
#print(df.isna().sum())

df_survived=df[df['Survived']==1]
df_not_survive=df[df['Survived']==0]
colors = ['gray', 'cyan']
labels = ['Deceased', 'Survived']

plt.hist([df_not_survive['Age'], df_survived['Age']], bins = 8, histtype='bar', rwidth=0.8, color = colors, label = labels )
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Количество выживших в зависимости от возраста')
plt.legend()
plt.show()

plt.hist([df_not_survive['Sex'], df_survived['Sex']], bins = 3, histtype='bar', rwidth=0.8, color = colors, label = labels )
plt.xlabel('Пол')
plt.ylabel('Count')
plt.title('Количество выживших в зависимости от пола')
plt.legend()
plt.show()

plt.hist([df_not_survive['Pclass'], df_survived['Pclass']], bins = 3, histtype='bar', rwidth=0.8, color = colors, label = labels )
plt.xlabel('Класс')
plt.ylabel('Count')
plt.title('Количество выживших в зависимости от класса пассажира')
plt.legend()
plt.show()
```
